```python
import math 
import nltk
from nltk.tokenize import word_tokenize
from nltk import FreqDist
from nltk.util import bigrams,trigrams,ngrams
```

```python
#function return vocab of corpus
def get_vocab(tokens):
    token_count=dict()
    for x in tokens:
        if x in token_count:
            token_count[x]+=1
        else:
            token_count[x]=1 
    return token_count
```

```python
#opening training corpus and calculating unnigram count and probability
from nltk.tokenize import word_tokenize

file = open('smoothing_text.txt')
text = file.read()
tokens=word_tokenize(text)
vocab = get_vocab(tokens)
print(vocab)
```

```python
def count_matrix(windo):
    n_gram=windo*2+1
    ngram = nltk.ngrams(tokens,n_gram)
    ngram_freq = nltk.FreqDist(ngram)

    matrix=dict()
    for key,value in ngram_freq.items():
        print(key,value,key[windo]);
        for i in range(0,windo):
#             print(key[i],value);
            pair=(key[windo],key[i])
#             print(pair)
            current_frequency = matrix.get(pair,0)
            matrix[pair]=current_frequency+value

        for i in range(windo+1,len(key)):
#             print(key[i],value);
            pair=(key[windo],key[i])
#             print(pair)
            current_frequency = matrix.get(pair,0)
            matrix[pair]=current_frequency+value
            
    return matrix
```

```python
window = int(input())
cont_matrix=count_matrix(window)
print(cont_matrix)
```

```python
def cosine_similarity(word_1,word_2):
    mag_w1=0
    mag_w2=0
    dot_pro=0
    for word in vocab:
        pair_1=(word_1,word);
        pair_2=(word_2,word);
        piar_1=cont_matrix.get(pair_1,0)
        piar_2=cont_matrix.get(pair_2,0)
        
        dot_pro = dot_pro + (piar_1*piar_2);
        
        mag_w1=mag_w1+piar_1*piar_1;
        mag_w2=mag_w2+piar_2*piar_2;
        
    mag_w1=math.sqrt(mag_w1)
    mag_w2=math.sqrt(mag_w2)
    cos_sim=dot_pro/(mag_w1*mag_w2)
    
    return cos_sim
```
